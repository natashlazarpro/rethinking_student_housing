# Colour Palettes Design

Выбор палитры для визуализации данных влючает в себя не только поиск гармоничных цветовб, но и построение логики по который эти цвета будут использованы. Задача по подбору цветовой палитры для исследования осложняеться необходимостью визуализировать картографические данные. 
Для визуалищации картографических данных используеться три вида цветовых схем:
* sequential schemes - mapping orderable or numeric data that progress from low to high (or vice versa)
* diverging schemes - mapping quantitative data above and below an interesting mid-point
* qualitative schemes - mapping categorical differences in qualitative data 

Специализированные вебсайты, такие как [ColorBrewer](https://colorbrewer2.org/#type=sequential&scheme=BuGn&n=3), [MapColPal](https://mapcolpal.github.io/?seed=68a5d5-aa90cf-d08599-c8915f-94a865-50b2a3) and [CartoColors](https://carto.com/carto-colors/) не только помогают в подборе всех трех видов цыетовых схем, но и наглядно демонтрируют как цвета будут смотреться на картах.

```python
import matplotlib.pyplot as plt
from matplotlib import colors

cols    = max(len(p) for p in color_palettes.values())
cols    = max(len(p) for p in color_palettes.values())
rows    = len(color_palettes)
fig, axs = plt.subplots(rows, 1, figsize=(cols*1, rows*1), constrained_layout=True)

if rows == 1:
    axs = [axs]

for ax, (name, palette) in zip(axs, color_palettes.items()):
    rgb = [colors.hex2color(c) for c in palette]
    arr = [rgb]                    

    ax.imshow(arr, aspect='auto')
    ax.set_xticks([]); ax.set_yticks([])
    ax.set_ylim(-0.40, 0)

    for i, c in enumerate(palette):
        ax.text(i, -0.5, c, ha='center', va='center',
                fontsize=8, color='black')

    ax.set_title(name, loc='left', fontsize=12, pad=10)
    ax.set_axis_off() 
```
